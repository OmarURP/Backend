```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.ticker import FormatStrFormatter

# Datos para la gráfica - usando LaTeX con saltos de línea
benchmarks = [r'$\mathrm{Agua\ residual}$ ' + '\n' + r'$\mathrm{colectada}$', 
              r'$\mathrm{Agua\ residual}$ ' + '\n' + r'$\mathrm{no\ colectada}$', 
              r'$\mathrm{Agua\ residual}$ ' + '\n' + r'$\mathrm{industrial}$', 
              r'$\mathrm{Actividad}$ ' + '\n' + r'$\mathrm{Pecuaria}$', 
              r'$\mathrm{Escurrimientos}$ ' + '\n' + r'$\mathrm{usos\ de\ suelo}$']
subtitulos = ['', '', '', '', '', '']

# Modelos y sus valores para cada benchmark
modelos = [r'$\mathrm{Valle\ de\ México}$', r'$\mathrm{Subcuenca\ Requena}$',  r'$\mathrm{Subcuenca\ Endhó}$']

# Datos de cada modelo para cada benchmark
HHG_1 = [ 45, 0, 0, 0, 0]
HHG_2 = [ 0.4, 0.1, 0.2, 4.2, 0.6]
HHG_3 = [ 0.8, 0.2, 1.1, 5.0, 0.3]

# Datos organizados para la gráfica
datos = [HHG_1, HHG_2, HHG_3]

# Configuración de la figura con 3 subplots para los 3 cortes
fig, (ax_top, ax_middle, ax_bottom) = plt.subplots(3, 1, figsize=(25, 15), sharex=True,
                                                   gridspec_kw={'height_ratios': [1, 1, 2], 'hspace': 0.03})

# Hacer transparente el fondo de la figura
fig.patch.set_alpha(0.0)
ax_top.patch.set_alpha(0.0)
ax_middle.patch.set_alpha(0.0)
ax_bottom.patch.set_alpha(0.0)

# Ancho de barra y separación
ancho_total = 0.8
ancho_barra = ancho_total / len(modelos)
indices = np.arange(len(benchmarks))

# Colores personalizados para cada modelo
colores = ['#3366FF', '#C0C0C0', '#99CCFF', '#DDDDDD', '#E6F0FF']

# Dibujar barras en los tres ejes
for i, benchmark_idx in enumerate(indices):
    valores_no_cero = [(j, datos[j][i]) for j in range(len(modelos)) if datos[j][i] > 0]
    num_barras = len(valores_no_cero)
    
    if num_barras == 0:
        continue
    
    ancho_total_barras = num_barras * ancho_barra
    
    for idx, (modelo_idx, valor) in enumerate(valores_no_cero):
        posicion = benchmark_idx - ancho_total_barras/2 + (idx + 0.5) * ancho_barra
        
        # Dibujar en los tres ejes (el recorte se hace con los límites)
        ax_top.bar(posicion, valor, width=ancho_barra, color=colores[modelo_idx],
                  edgecolor='black', linewidth=0.5)
        ax_middle.bar(posicion, valor, width=ancho_barra, color=colores[modelo_idx],
                     edgecolor='black', linewidth=0.5)
        ax_bottom.bar(posicion, valor, width=ancho_barra, color=colores[modelo_idx],
                     edgecolor='black', linewidth=0.5)

# Configurar límites para crear el efecto de 3 cortes
ax_top.set_ylim(15, 60)     # Parte superior para el valor de 45
ax_middle.set_ylim(3.5, 6.5)    # Parte media para valores de 4-6
ax_bottom.set_ylim(0, 1.5)    # Parte inferior para valores de 0-1

# Ocultar spines para crear el efecto de continuidad
ax_top.spines['bottom'].set_visible(False)
ax_middle.spines['top'].set_visible(False)
ax_middle.spines['bottom'].set_visible(False)
ax_bottom.spines['top'].set_visible(False)
ax_top.spines['right'].set_visible(False)
ax_middle.spines['right'].set_visible(False)
ax_bottom.spines['right'].set_visible(False)

# Configurar ticks y formatear a 2 decimales
ax_top.tick_params(bottom=False, labelbottom=False)
ax_middle.tick_params(bottom=False, labelbottom=False, top=False, labeltop=False)

# Formatear todos los ejes Y a 2 decimales
ax_top.yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
ax_middle.yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
ax_bottom.yaxis.set_major_formatter(FormatStrFormatter('%.1f'))

# Configurar tamaño de fuente para los ticks
ax_top.tick_params(axis='y', labelsize=25)
ax_middle.tick_params(axis='y', labelsize=25)
ax_bottom.tick_params(axis='y', labelsize=25)

# Añadir líneas diagonales para mostrar los recortes
d = 0.00
kwargs = dict(transform=ax_top.transAxes, color='k', clip_on=False, linewidth=2)

# Líneas en el subplot superior
ax_top.plot((-d*2, 0), (-d*2, +d*2), **kwargs)        # extremo izquierdo superior
ax_top.plot((1, 1+d*2), (-d*2, +d*2), **kwargs)       # extremo derecho superior

# Líneas en el subplot medio
kwargs.update(transform=ax_middle.transAxes)
ax_middle.plot((-d*2, 0), (1-d*2, 1+d*2), **kwargs)   # líneas superiores del medio
ax_middle.plot((1, 1+d*2), (1-d*2, 1+d*2), **kwargs)
ax_middle.plot((-d*2, 0), (-d*2, +d*2), **kwargs)     # líneas inferiores del medio
ax_middle.plot((1, 1+d*2), (-d*2, +d*2), **kwargs)

# Líneas en el subplot inferior
kwargs.update(transform=ax_bottom.transAxes)
ax_bottom.plot((-d*2, 0), (1-d*2, 1+d*2), **kwargs)   # extremo izquierdo inferior
ax_bottom.plot((1, 1+d*2), (1-d*2, 1+d*2), **kwargs)  # extremo derecho inferior

# Configurar grid
ax_top.grid(axis='y', linestyle='--', alpha=0.3)
ax_middle.grid(axis='y', linestyle='--', alpha=0.3)
ax_bottom.grid(axis='y', linestyle='--', alpha=0.3)

# Configurar etiquetas del eje X solo en el gráfico inferior
ax_bottom.set_xticks(indices)
ax_bottom.set_xticklabels(benchmarks, fontsize=25, ha='center')

# Etiqueta del eje Y (centrada entre los tres subplots)
fig.text(0.04, 0.5, r'$\mathrm{HHG\ por\:NT\:(km^3 \cdot año^{-1})}$', 
         va='center', rotation='vertical', fontsize=25)

# Configurar estilos de barras para la leyenda
elementos_leyenda = []
for modelo, color in zip(modelos, colores):
    elementos_leyenda.append(Patch(facecolor=color, edgecolor='black', label=modelo))

# Añadir leyenda
fig.legend(handles=elementos_leyenda, loc='lower center', bbox_to_anchor=(0.5, -0.02),
          ncol=len(modelos), frameon=False, fontsize=30)

# Añadir subtítulos debajo de los nombres de benchmarks
for i, (benchmark, subtitulo) in enumerate(zip(benchmarks, subtitulos)):
    ax_bottom.text(i, -0.3, subtitulo, ha='center', fontsize=10)

# Ajustar el layout
plt.subplots_adjust(bottom=0.15)

# Guardar la figura con fondo transparente
plt.savefig('comparativa_modelos.png', dpi=300, bbox_inches='tight', transparent=True)

# Mostrar gráfica
plt.show()
```
