```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Datos para la gráfica
#benchmarks = ['Agua residual \n colectada', 'Agua residual \n no colectada', 'Agua residual \n industrial', 'Actividad \n Pecuaria', 'Escurrimientos \n usos de suelo']
# Datos para la gráfica - usando LaTeX con saltos de línea
benchmarks = [r'${Agua\ residual}$ ' + '\n' + r'${colectada}$', 
              r'${Agua\ residual}$ ' + '\n' + r'${no\ colectada}$', 
              r'${Agua\ residual}$ ' + '\n' + r'${industrial}$', 
              r'${Actividad}$ ' + '\n' + r'${Pecuaria}$', 
              r'${Escurrimientos}$ ' + '\n' + r'${usos\ de\ suelo}$']
subtitulos = ['', '', '', '', '', '']

# Modelos y sus valores para cada benchmark
modelos = [r'$Valle\ de\ México$', r'$Subcuenca\ Requena$',  r'$Subcuenca\ Endhó$']

# Datos de cada modelo para cada benchmark
HHG_1 = [ 45, 0, 0, 0, 0]
HHG_2 = [ 0.4, 0.1, 0.2, 4.2, 0.6]
HHG_3 = [ 0.8, 0.2, 1.1, 5.0, 0.3]
#openai_o1_mini = [63.6, 93.4, 60.0, 90.0, 85.2, 41.6]
#deepseek_v3 = [39.2, 58.7, 59.1, 90.2, 88.5, 42.0]

# Datos organizados para la gráfica
datos = [HHG_1, HHG_2, HHG_3]

# Configuración de la figura
fig, ax = plt.subplots(figsize=(25, 12))
plt.subplots_adjust(bottom=0.15)

# Hacer transparente el fondo de la figura y el área del gráfico
fig.patch.set_alpha(0.0)
ax.patch.set_alpha(0.0)

# Ancho de barra y separación
ancho_total = 0.8
ancho_barra = ancho_total / len(modelos)
indices = np.arange(len(benchmarks))

# Colores personalizados para cada modelo
colores = ['#3366FF', '#C0C0C0', '#99CCFF', '#DDDDDD', '#E6F0FF']

 
# Configurar estilos de barras para la leyenda
elementos_leyenda = []
for modelo, color in zip(modelos, colores):
    elementos_leyenda.append(Patch(facecolor=color, edgecolor='black', label=modelo))

# Añadir leyenda en la parte superior
ax.legend(handles=elementos_leyenda, loc='lower center', bbox_to_anchor=(0.5, -0.3),
          ncol=len(modelos), frameon=False, fontsize=30)

# Configurar ejes y etiquetas
ax.set_xticks(indices)
ax.set_xticklabels(benchmarks, fontsize=25)

for i, benchmark_idx in enumerate(indices):
    # Filtrar modelos con valores mayores que 0 para este benchmark
    valores_no_cero = [(j, datos[j][i]) for j in range(len(modelos)) if datos[j][i] > 0]
    num_barras = len(valores_no_cero)
    
    if num_barras == 0:
        continue  # Si no hay valores, saltar este benchmark
    
    # Calcular el ancho total ocupado por las barras presentes
    ancho_total_barras = num_barras * ancho_barra
    
    # Posicionar y dibujar solo las barras con valores, centradas
    for idx, (modelo_idx, valor) in enumerate(valores_no_cero):
        # Calcular posición centrada en el benchmark
        posicion = benchmark_idx - ancho_total_barras/2 + (idx + 0.5) * ancho_barra
        barra = ax.bar(posicion, valor, width=ancho_barra, color=colores[modelo_idx],
                     edgecolor='black', linewidth=0.5)
        
        # Añadir etiqueta de valor
        #if valor > 5:  # Solo mostrar etiqueta si el valor es significativo
        #    ax.text(posicion, valor + 1, f'{valor}', ha='center', va='bottom', 
        #           fontsize=20)

# Configurar límites y etiquetas de los ejes
ax.set_ylim(0, 50)
ax.set_ylabel(r'$HHG\ por\:NT\:(km^3 \cdot año^{-1})$', fontsize=25)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Eliminar bordes superiores y laterales
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
# Configurar ejes y etiquetas
ax.set_xticks(indices)
ax.set_xticklabels(benchmarks, fontsize=25, ha='center')
plt.yticks(fontsize=25)

# ...existing code...

# Añadir subtítulos debajo de los nombres de benchmarks
for i, (benchmark, subtitulo) in enumerate(zip(benchmarks, subtitulos)):
    ax.text(i, -5, subtitulo, ha='center', fontsize=10)

# Añadir líneas de cuadrícula horizontales
plt.grid(axis='y', linestyle='--', alpha=0.3)


# Guardar la figura con fondo transparente
plt.savefig('comparativa_modelos.png', dpi=300, bbox_inches='tight', transparent=True)

# Mostrar gráfica
plt.show()
```
